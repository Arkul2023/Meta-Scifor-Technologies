In reallife scenarios we dont have multiple different training set on which we can train our model separately and at the end  combine their result . Here , bootstrapping comes into picture . BootStrapping is a technique of sampling different sets of data from a given training set by using replacement . After bootstraping the training dataset . we train model on all the different sets and aggregate the resul . The technique is known as BootStrap Aggregation or Bagging.






Bagging - Bagging is the type of ensembles technique in which a single training algorithm is used on different subsets of the training data where the subsets of the training data where the subset sampling is done with replacemennt (bootstrap) . once the algorithmis trained on all the subsets . then bagging makes the prediction by aggreagating all the prediction made by the algorithm on different subsets. In case of regression bagging prediction is simply the mean of all the prediction and in the case of classifer bagging prediction is the most frequent prediction (Majority Vote) among all the predictions.




Bagging is also know as PARALLEL MODEL Since we run all models parallely and combine there result at the end .

# Python Implementation

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier


import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
dataset=load_breast_cancer()

x=dataset.data
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4)
```

```python
knn=DecisionTreeClassifier()
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

# Let Use Bagging Over KNN and see if our score improves.

```python
bag_knn=BaggingClassifier(knn,n_estimators=9,max_samples=0.7,bootstrap=True,oob_score=True)
```

```python
bag_knn.fit(x_train,y_train)
bag_knn.score(x_test,y_test)
```

- Great Our Score Significantly improved with use of bagging.



- Let not use bootstarp and see the model accuracy ! Remember this is pasting. 

```python
pasting_knn= BaggingClassifier(knn,n_estimators=9,max_samples=0.7,bootstrap=False,random_state=3)
```

```python
pasting_knn.fit(x_train,y_train)
pasting_knn.score(x_test,y_test)
```

# Split into Train Test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=94255)
```

# Model Confidence/Accuracy

```python
def metrics_score(cnn,x_train,x_test,y_train,y_test,train=True):
    if train:
        y_pred=cnn.predict(x_train)
        print("=================Training Score===============")
        print("Accuracy Score ==> ",accuracy_score(y_train,y_pred)*100)
    elif train==False:
        pred=cnn.predict(x_test)
        print("=================Testing Score===============")
        print("Accuracy Score ==> ",accuracy_score(y_test,pred)*100)
        print("Classification Report==> ", classification_report(y_test,pred))
```

# Initiate DecisionTreeClassifier with new parameter and Train 

```python
random_clf =RandomForestClassifier()
```

```python
#Train the model
random_clf.fit(x_train,y_train)
```

```python
#Call the fucntion and pass dataset to check train and test score
```

```python
metrics_score(random_clf,x_train,x_test,y_train,y_test,train=True)
metrics_score(random_clf,x_train,x_test,y_train,y_test,train=False)
```

So , Our Model is Overfitted . Using GridSearchCV we will tune our parameter so here what parameter you can use . So this random Forest will use internally Decision Tree by default .

# Lets tune the parameter using GridSearchCV

```python
#Random Forest Classifier 
params={'n_estimators':[13,15],
       'criterion':['entropy','gini'],
       'max_depth':[10,15],
       'min_samples_split':[10,13],
       'min_samples_leaf':[5,8]}

grd=GridSearchCV(random_clf,param_grid=params)
grd.fit(x_train,y_train)

print("Best_Params ==> ",grd.best_params_)
```

```python
random_clf=grd.best_estimator_ #reinstantiating with best params
random_clf.fit(x_train,y_train)
```

```python
metrics_score(random_clf,x_train,x_test,y_train,y_test,train=True)
metrics_score(random_clf,x_train,x_test,y_train,y_test,train=False)
```

# Lets Plot ROC AUC Curve ti choose best model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
lr=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
```

```python
#Train Test Split
```

```python
x=df.drop(columns=['cardio'],axis=1)
y=df.cardio
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=94062)
```

```python
#Training with all classifiers
lr.fit(x_train,y_train)
rf.fit(x_train,y_train)
knn.fit(x_train,y_train)
dt.fit(x_train,y_train)
print("All Models are Trained")
```

# Lets Check ROC AUC Curve fro the fitted Model

- True Positive Rate(TPR)=Out OF 100 actual cardio patients how many our model predicted as cardio patients correctly (Higher The Better)


- False Positive Rate (FPR)=How many people were predicted as cardio patients but in reality they were healthy (Lower the Better)





Since it is difficult to evaluate the score based on only TPR and FPR data. There comes AUC.

```python
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt

# Plot ROC curves for all models
ax = plt.gca()  # Get current axis
RocCurveDisplay.from_estimator(dt, x_train, y_train, name="Decision Tree", ax=ax)
RocCurveDisplay.from_estimator(lr, x_train, y_train, name="Logistic Regression", ax=ax)
RocCurveDisplay.from_estimator(knn, x_train, y_train, name="KNN", ax=ax)
RocCurveDisplay.from_estimator(rf, x_train, y_train, name="Random Forest", ax=ax)
plt.legend(prop={'size': 10}, loc='lower right')
plt.show()
```
