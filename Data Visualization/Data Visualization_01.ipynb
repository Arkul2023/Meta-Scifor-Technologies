```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline                          
from numpy.random import randn
```

```python
## Here '%' is used to plot the graphs inline.
## Here in last line we import random module from numpy as randn which sample normal distribution
```

```python
df = pd.DataFrame(randn(10,4),columns=['a', 'b', 'c', 'd'])
```

```python
df.tail()
```

```python
## A random table is created
```

```python
## Plots the graph of the above table
```

```python
df.plot(kind='bar')
plt.show()
```

```python
df.plot.bar()
plt.show()
```

```python
df.plot.barh()
plt.show()
```
